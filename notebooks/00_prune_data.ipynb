## plants_gbif/occurrence

```python
in_file = DATA / 'plants_gbif' / 'occurrence.txt.gz'
out_file = PRUNED / 'plants_gbif_occurrence.csv.gz'
```

```python
headers = !zcat "$in_file" | head -1
headers = headers[0].split('\t')
headers
```

```python
reader = pd.read_csv(
    in_file, sep='\t', dtype=str, keep_default_na=False,
    chunksize=CHUNK,
    usecols=['gbifID',
             'decimalLatitude', 'decimalLongitude',
             'verbatimCoordinateSystem',
             'coordinateUncertaintyInMeters', 'coordinatePrecision',
             'locality', 'verbatimLocality'])

first_chunk = True
for df in tqdm(reader):
    if first_chunk:
        df.to_csv(out_file, index=False)
    else:
        df.to_csv(out_file, index=False, mode='a', header=False)
    first_chunk = False
```
